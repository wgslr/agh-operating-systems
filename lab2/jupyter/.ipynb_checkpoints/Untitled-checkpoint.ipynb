```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(filepath_or_buffer="wyniki.txt",
                   sep="[;\:\(]",
                   skipinitialspace=True,
                   engine="python",
                   header=None,
                   names = ["operation", "records", "rsize", "syscalls", "kind", "_", "time"])

# strip whitespace
strings = data.select_dtypes(['object'])
data[strings.columns] = strings.apply(lambda x: x.str.strip())

data['syscalls'] = data.syscalls == "sys"
data['bytes'] = data.records * data.rsize
data['speed'] = data.bytes / data.time
data
```

```python
data
```

```python
#display(data[data.kind == ' user '])

fig, ax = plt.subplots()

plotted1 = data[True == data.syscalls][data.kind == ' real '][['size', 'speed']].set_index('size').sort_index()
plotted2 = data[False == data.syscalls][data.kind == ' real '][['size', 'speed']].set_index('size').sort_index()
#plotted.sort_index
#display(plotted)
#display(plotted1)
#display(plotted2)

ax = plt.plot(plotted1, label='sys')
ax = plt.plot(plotted2, label='lib')
plt.legend(loc='best')
plt.title('bytes per ms')


plt.show()
```
