```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Times.all.txt', sep=";")
```

```python
data
```

```python
#data.plot(x = 'filter', y='time',)
for t in [1,2,4,8]:
    plt.plot("filter", "time", label=t, data=data[data.threads==t])
plt.legend()
plt.xlabel("filter area")
plt.ylabel("time [ms]")
plt.show()
```

```python
data['operations'] = data.img * data['filter']
data['opms'] = data.operations / data.time
data
```
